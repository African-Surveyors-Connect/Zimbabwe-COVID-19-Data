```python
import pandas as pd
from arcgis.features import GeoAccessor, GeoSeriesAccessor
```

## Dataset Feature Identifications

Here is a list of all thew Feature ID which can be used for performing the data queries from the Feature Layer (Hosted) within the **COVID-19 Hub for Zimbabwe**. 

*Infections Time Series* - `42ec33c9361d49b585a23d780207726d`

*Vaccination Time Series* - `0bad4380917a48da8f4f12028709c443`. 
The Vaccination Times Series contains 3 layers within the Feature Layer. For the Time Series Data query the layer with Index [2]

*Provincial Cases* - `122efe4c5ab54ff9a9e75cc1908d48f4`

*Provincial Time Series* - `20703dd3a24f45f08ea37034285d3492`. 
This one is a new introduction to the daily reporting system for COVID-19 within the Hub. 

```python
from arcgis import GIS
gis = GIS()
item = gis.content.get("42ec33c9361d49b585a23d780207726d") # Contains the Feature ID for the Infections Time-Series layer
flayer = item.layers[0]

# create a Spatially Enabled DataFrame object
sdf = pd.DataFrame.spatial.from_layer(flayer)
sdf.head()
```

## Query The Layer to Return intended Display

Lets return the records for the latest data update for out data

```python
''' retrieve using the largest number of Cumulative tests
Would really love to use the date but its a bit buggy
'''
largest_cum_value = sdf["Total_Cumu"].max()

# Let's see if it works
print(largest_cum_value)
```

```python
latest_records = sdf.query("Total_Cumu == @largest_cum_value")
```

```python
# Retrieve the Latest Records from the data
latest_records[['Date', 'Total_Cumu', 'New_Cases', 'PCR', 'Recovered', 'Deaths', 'Daily_Posi', 'conditio_1', 'conditio_2', 'asymptomat', 'total_vacc', 'total_vacc_2', 'hotspot_cities', 'lockdown_update', 'keyNotes', 'newHosp']].head()
```
